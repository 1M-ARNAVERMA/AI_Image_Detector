```python
import os, random, shutil

BASE = "../dataset"
OUTPUT = "../processed_data"

SPLITS = ["train", "val", "test"]
CATEGORIES = ["real", "ai", "edited"]

for split in SPLITS:
    for cat in CATEGORIES:
        os.makedirs(f"{OUTPUT}/{split}/{cat}", exist_ok=True)

for cat in CATEGORIES:
    files = os.listdir(f"{BASE}/{cat}")
    random.shuffle(files)

    train = files[:int(0.7*len(files))]
    val = files[int(0.7*len(files)):int(0.85*len(files))]
    test = files[int(0.85*len(files)):]

    for f in train:
        shutil.copy(f"{BASE}/{cat}/{f}", f"{OUTPUT}/train/{cat}/{f}")

    for f in val:
        shutil.copy(f"{BASE}/{cat}/{f}", f"{OUTPUT}/val/{cat}/{f}")

    for f in test:
        shutil.copy(f"{BASE}/{cat}/{f}", f"{OUTPUT}/test/{cat}/{f}")
```
