```python
import numpy as np
from data_loader import load_images
from feature_extractor import extract_features

X_train, y_train = load_images("../processed_data/train")

features = []
for img in X_train:
    features.append(extract_features((img*255).astype("uint8")))

X_train_features = np.array(features)

np.save("../results/X_train.npy", X_train_features)
np.save("../results/y_train.npy", y_train)

print("Feature extraction complete")
```
