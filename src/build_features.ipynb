```python
import os
import cv2
import numpy as np

BASE_PATH = "../processed_data/train"
CATEGORIES = ["real", "ai", "edited"]
IMG_SIZE = 224

X_features = []
y_labels = []

for label, category in enumerate(CATEGORIES):
    folder = os.path.join(BASE_PATH, category)

    for img_name in os.listdir(folder):
        img_path = os.path.join(folder, img_name)

        img = cv2.imread(img_path)
        if img is None:
            continue

        img = cv2.resize(img, (IMG_SIZE, IMG_SIZE))
        features = extract_features(img)

        X_features.append(features)
        y_labels.append(label)

    print(f"Processed category: {category}")

X_features = np.array(X_features, dtype=np.float32)
y_labels = np.array(y_labels)

np.save("../results/X_train.npy", X_features)
np.save("../results/y_train.npy", y_labels)

print("✅ Feature extraction completed safely")
```
