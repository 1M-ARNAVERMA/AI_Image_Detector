```python
import numpy as np
import joblib
from sklearn.metrics import classification_report, confusion_matrix

# Load trained models
pca = joblib.load("../models/pca_model.pkl")
svm = joblib.load("../models/svm_model.pkl")

# Load validation data
X_val = np.load("../results/X_val.npy")
y_val = np.load("../results/y_val.npy")

# Apply PCA learned from training data
X_val_pca = pca.transform(X_val)

# Predict
y_pred = svm.predict(X_val_pca)
y_prob = svm.predict_proba(X_val_pca)

print("Validation Classification Report:")
print(classification_report(y_val, y_pred))

print("Confusion Matrix:")
print(confusion_matrix(y_val, y_pred))

print("Sample probability output:", y_prob[0])
```
